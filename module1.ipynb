```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("education_career_success.csv")
df
```

```python
internshipOffers = df.groupby("Internships_Completed")["Job_Offers"].mean()
plt.bar(internshipOffers.index, internshipOffers.values)
plt.title("Average Job Offers based on Internships")
plt.xlabel("Internships Completed")
plt.ylabel("Average Number of Job Offers")
```

```python
certificationOffers = df.groupby("Certifications")["Job_Offers"].mean()
plt.bar(certificationOffers.index, certificationOffers.values, color="orange")
plt.title("Average Job Offers based on Certifications")
plt.xlabel("Certifications Completed")
plt.ylabel("Average Number of Job Offers")
```

```python
internshipSalaries = df.groupby("Internships_Completed")["Starting_Salary"].mean()
plt.bar(internshipSalaries.index, internshipSalaries.values)
plt.title("Average Starting Salary Based on Internships Completed")
plt.xlabel("Internships Completed")
plt.ylabel("Average Starting Salary ($)")
```

```python
certificationSalaries = df.groupby("Certifications")["Starting_Salary"].mean()
plt.bar(certificationSalaries.index, certificationSalaries.values, color="orange")
plt.title("Average Starting Salary Based on Certifications Completed")
plt.xlabel("Certifications Completed")
plt.ylabel("Average Starting Salary ($)")
```

```python
grouped = df.groupby(["Field_of_Study", "Internships_Completed"])["Job_Offers"].mean().unstack()
grouped.T.plot(kind="bar", figsize=(14, 7))
plt.title("Average Job Offers by Internships for Each Field of Study")
plt.xlabel("Internships Completed")
plt.ylabel("Average Job Offers")
```

```python
df.boxplot(column="Starting_Salary", by="Internships_Completed")
plt.title("Starting Salary Based on Internships Completed")
plt.suptitle("")
plt.xlabel("Internships Completed")
plt.ylabel("Starting Salary ($)")
```

```python
df.boxplot(column="Starting_Salary", by="Certifications")
plt.title("Starting Salary Based on Certifications Completed")
plt.suptitle("")
plt.xlabel("Certifications Completed")
plt.ylabel("Starting Salary ($)")
```
